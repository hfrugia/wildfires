Quantitative Assessment

```python
# Load NIR and SWIR bands
# Landsat 8: NIR = Band 5, SWIR2 = Band 7
prefire_nir  = Raster("prefire_B5.tif")
prefire_swir = Raster("prefire_B7.tif")
postfire_nir  = Raster("midfire_B5.tif")
postfire_swir = Raster("midfire_B7.tif")

# --- Step 1: Calculate NBR for prefire and postfire
nbr_prefire = (prefire_nir - prefire_swir) / (prefire_nir + prefire_swir)
nbr_postfire = (postfire_nir - postfire_swir) / (postfire_nir + postfire_swir)

# --- Step 2: Calculate dNBR
dnbr = nbr_prefire - nbr_postfire

# Optional: Save the dNBR raster if you want to inspect it
dnbr_output = os.path.join(arcpy.env.workspace, "dnbr.tif")
dnbr.save(dnbr_output)

# Add to current ArcGIS Pro map
aprx = arcpy.mp.ArcGISProject("CURRENT")
map_obj = aprx.listMaps()[0]
map_obj.addDataFromPath(dnbr_output)

print("✅ dNBR calculated and added to the map.")
```

```python
# Reclassify
remap = RemapRange([
    [0.1, 0.27, 1],
    [0.27, 0.44, 2],
    [0.44, 0.66, 3],
    [0.66, 1.0, 4]
])
burn_severity = Reclassify(dnbr_clipped, "VALUE", remap)

# Save output
burn_severity_path = r"C:\Users\hfrug\OneDrive\Documents\Bootcamp GIS\Wildfire Assessments\Data\California_Wildfires_Landsat\burn_severity.tif"
burn_severity.save(burn_severity_path)

input_folder = r"C:\Users\hfrug\OneDrive\Documents\Bootcamp GIS\Wildfire Assessments\Data\California_Wildfires_Landsat"
clr_path = f"{input_folder}\\burn_colormap.clr"

with open(clr_path, "w") as f:
    f.write("1 230 0 0\n")    # Red
    f.write("2 242 85 0\n")   # Orange
    f.write("3 250 142 0\n")  # Orange-yellow
    f.write("4 255 195 0\n")  # Yellow

# Apply the colormap
arcpy.management.AddColormap(burn_severity_path, None, clr_path)

print("✅ Final burn severity layer saved and symbolized with 4 colors.")
```

Area calculation (local)

```python
# Get pixel size directly
pixel_size = burn_severity.meanCellWidth  # assume square pixels
pixel_area = pixel_size * pixel_size  # in sq meters

# Get raster as numpy array
burn_array = arcpy.RasterToNumPyArray(burn_severity)

# Count pixels per class
counts_per_class = []
for burn_class in [1, 2, 3, 4]:
    count = np.sum(burn_array == burn_class)
    counts_per_class.append(count)

# Sum and calculate area
total_pixels = sum(counts_per_class)
sqm_area = total_pixels * pixel_area
acres = sqm_area * 0.00024711

print(f"Total burned area: {sqm_area:,.2f} sq meters (~{acres:.2f} acres)")
```

Raster to Feature layer conversion

```python
# Define the path to your raster file
in_raster = r"C:\Users\hfrug\OneDrive\Documents\Bootcamp GIS\Wildfire Assessments\Data\California_Wildfires_Landsat\burn_severity.tif"

# Define output path and name for the new feature class (e.g., shapefile or GDB feature class)
out_features = r"C:\Users\hfrug\OneDrive\Documents\Bootcamp GIS\Wildfire Assessments\Data\California_Wildfires_Landsat\burn_severity_polygons.shp"
```

```python
# Convert raster to polygon
arcpy.RasterToPolygon_conversion(
    in_raster=in_raster,
    out_polygon_features=out_features,
    simplify="NO_SIMPLIFY",
    raster_field="VALUE"
)

# Add and calculate area
arcpy.management.AddField(out_features, "AreaSqM", "DOUBLE")
arcpy.management.CalculateGeometryAttributes(
    out_features,
    [["AreaSqM", "AREA"]],
    area_unit="SQUARE_METERS"
)

# Remove small polygons
threshold = 4500
with arcpy.da.UpdateCursor(out_features, ["AreaSqM"]) as cursor:
    for row in cursor:
        if row[0] < threshold:
            cursor.deleteRow()
```

```python
# Access current ArcGIS project and map
aprx = arcpy.mp.ArcGISProject("CURRENT")
map_obj = aprx.listMaps()[0]  # You can also use aprx.listMaps("Map")[0] if named

# Access layer
burn_severity_polygons = map_obj.listLayers("burn_severity_polygons")[0]

# Get data source path
input_fc = burn_severity_polygons.dataSource

# Set output path
output_buffer = r"C:\Users\hfrug\OneDrive\Documents\ArcGIS\Projects\California Fire Analysis\Outputs\CaliforniaFireBuffer.shp"

# Create buffer (100 meters, dissolve all, multipart)
arcpy.analysis.Buffer(
    in_features=input_fc,
    out_feature_class=output_buffer,
    buffer_distance_or_field="100 Meters",
    line_side="FULL",
    line_end_type="ROUND",
    dissolve_option="ALL",
    dissolve_field=[],
    method="PLANAR"
)

print("Buffer created at:", output_buffer)
```

```python
# Grab the relevant field names from the code_map
pop_columns = list(code_map.values())

# Select only those columns from the impacted census tracts
impacted_pop = impacted_sdf[pop_columns]

# Sum across all tracts
total_pop = impacted_pop.sum().reset_index()
total_pop.columns = ['field_name', 'population_count']

# Map field names back to readable names
total_pop['age_group'] = total_pop['field_name'].map({v: k for k, v in code_map.items()})

# Optional: sort by age group (if needed)
total_pop = total_pop[['age_group', 'population_count']]
```
